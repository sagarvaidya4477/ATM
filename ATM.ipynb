```python
balance = 1000
pin = 4477

print("Welcome to the ATM")
print("Please insert your card")

while True:
    entered_pin = int(input("Enter your 4-digit PIN: "))
    if entered_pin == pin:
        print("Welcome")
        while True:
            print("Choose the type of transaction you want to perform")
            print("1. Check Balance")
            print("2. Deposit Money")
            print("3. Withdraw Balance")
            print("4. Change PIN")
            print("5. Exit")
            choice = int(input("Enter your choice: "))
            
            if choice == 1:
                print("Your account balance is", balance)
            elif choice == 2:
                amount = int(input("Enter amount to deposit: "))
                if amount > 0:
                    balance += amount
                    print(amount, "has been deposited. Your new balance is:", balance)
                else:
                    print("Please enter a valid amount.")
            elif choice == 3:
                amount = int(input("Enter amount to withdraw: "))
                if 0 < amount <= balance:
                    balance -= amount
                    print(amount, "has been withdrawn. Your new balance is:", balance)
                else:
                    print("Please enter a valid amount.")
            elif choice == 4:
                old_pin = int(input("Enter your old PIN: "))
                if old_pin == pin:
                    new_pin = int(input("Enter your new PIN: "))
                    pin = new_pin
                    print("PIN has been changed successfully.")
                else:
                    print("Incorrect old PIN.")
            elif choice == 5:
                print("Thank you for using the ATM. Goodbye!")
                break
            else:
                print("Invalid choice. Please try again.")
        break
    else:
        print("Wrong PIN, please try again.")
```

```python
class Atm:
    def __init__(self):
        self.__pin = ""
        self.__balance = 0
        self.menu()

    def menu(self):
        print("Welcome")
        user_input = input("""
        1. Press 1 to create PIN
        2. Press 2 to change PIN
        3. Press 3 to check balance
        4. Press 4 to withdraw
        5. press 5 to Deposit Money
        6. Press 6 to exit
        """)
        if user_input == "1":
            self.create_pin()
            self.menu()
        elif user_input == "2":
            self.change_pin()
            self.menu()
        elif user_input=="3":
            self.check_balance()
            self.menu()
        elif user_input=="4":
            self.withdraw()
            self.menu()
        elif user_input=="5":
            self.Deposit_Money()
            self.menu()
        elif user_input=="6":
            print("Thank you")
        else:
            print("invalid option")
    def create_pin(self):
        self.__pin = input("Enter your new PIN: ")
        print("PIN created successfully.")

    def change_pin(self):
        old_pin = input("Enter your old PIN: ")
        if old_pin == self.__pin:
            self.__pin = input("Enter your new PIN: ")
            print("PIN changed successfully.")
        else:
            print("Incorrect old PIN. Please try again.")
    def check_balance(self):
        entered_pin = input("Enter your PIN: ")
        if entered_pin == self.__pin:
            print("your balance is",self.__balance)
        else:
            print("please enter correct pin")
    def withdraw(self):
        entered_pin = input("Enter your PIN: ")
        if entered_pin == self.__pin:
            amount = int(input("Enter the amount to withdraw: "))
            if amount <= self.__balance:
                self.__balance -= amount
                print("Amount withdrawn successfully.")
                print("Your new balance is", self.__balance)
            else:
                print("low balance")
        else:
            print("Incorrect PIN. Please try again.") 
    def Deposit_Money(self):
        entered_pin=input("Enter your pin:")
        if entered_pin==self.__pin:
            self.__balance = int(input("Enter your initial balance: "))
            print("Your amount deposit successfully..now your current balance is:",self.__balance)
        else:
            print("Incorrect pin..please try again")
        
a1 = Atm()
```
